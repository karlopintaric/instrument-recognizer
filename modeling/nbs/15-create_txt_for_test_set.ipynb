```python
import pandas as pd
from sklearn.model_selection import StratifiedGroupKFold
import numpy as np
```

```python
df = pd.read_csv("../metadata_test.csv")
df.head()
```

```python
df.shape[0]
```

```python
kf = StratifiedGroupKFold(n_splits=2, shuffle=True, random_state=1)
splits = kf.split(df.fname, df.instr, groups=df.song_name)
valid, test = list(splits)[0]
```

```python
len(valid), len(test)
```

```python
df.iloc[valid].label_count.value_counts()
```

```python
df.iloc[test].label_count.value_counts()
```

```python
test_songs = df.iloc[test].fname.sort_values().to_numpy()
```

```python
with open('../test_songs.txt', 'w') as f:
    # iterate over the list of names and write each one to a new line in the file
    for song in test_songs:
        f.write(song + '\n')
```
